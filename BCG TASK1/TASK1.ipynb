```python
import pandas as pd
```

```python
df = pd.read_csv('First_Report_BCG.csv')
```

```python
df
```

Calculating Year-by-Year growth rates for Total Revenue and Net Income

```python
df['Revenue Growth (%)'] = df.groupby(['Company_Name'])['Total_Revenue'].pct_change() * 100
df['Net Income Growth (%)'] = df.groupby(['Company_Name'])['Net_Income'].pct_change() * 100
```

```python
df
```

Filling Null/NaN values that result from pct_change calculations with 0

```python
df.fillna(0,inplace=True)
```

```python
df
```

Calculating Year-by-Year growth rates for Total Assets, Total Liabilities and Cash flow from Operations Activities

```python
df['Assets Growth (%)'] = df.groupby('Company_Name')['Total_Assets'].pct_change() * 100
df['Liabilities Growth (%)'] = df.groupby('Company_Name')['Total_Liabilities'].pct_change() * 100
df['Cash Flow from Operations Growth(%)'] = df.groupby('Company_Name')['Cash_Flow'].pct_change() * 100
```

```python
df
```

```python
df.fillna(0,inplace=True)
```

```python
df
```

```python
df.to_csv('final_BCG_report.csv')
```

```python
summary = df.groupby('Company_Name').agg({
    'Revenue Growth (%)': 'mean',
    'Net Income Growth (%)': 'mean',
    'Assets Growth (%)' : 'mean',
    'Liabilities Growth (%)' : 'mean',
    'Cash Flow from Operations Growth(%)' :'mean'    
}).reset_index()
```
